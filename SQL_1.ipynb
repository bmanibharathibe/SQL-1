```python



import sqlite3

# Create connection
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

# Create table
cursor.execute("""
CREATE TABLE employees (
    employee_id INTEGER PRIMARY KEY,
    name TEXT,
    department TEXT,
    salary INTEGER,
    hire_date TEXT
)
""")

# Insert data
cursor.executemany("""
INSERT INTO employees VALUES (?, ?, ?, ?, ?)
""", [
(1,'John','Engineering',120000,'2020-01-15'),
(2,'Priya','Finance',95000,'2019-03-10'),
(3,'Rahul','Sales',80000,'2021-06-20'),
(4,'Meena','Marketing',70000,'2022-02-12'),
(5,'Arjun','HR',75000,'2018-09-25'),
(6,'Kiran','Engineering',88000,'2023-01-05'),
(7,'Ravi','Marketing',85000,'2021-08-10'),
(8,'Sneha','Finance',99000,'2020-11-18'),
(9,'Vikram','Operations',72000,'2017-07-07'),
(10,'Anita','Sales',91000,'2022-03-22')
])

conn.commit()

# Function to run and print query
def run_query(title, query):
    print("\n", title)
    cursor.execute(query)
    for row in cursor.fetchall():
        print(row)


# Task 1

run_query("Task 1 - Marketing Employees",
          "SELECT * FROM employees WHERE department = 'Marketing'")

run_query("Task 1 - Salary > 90000",
          "SELECT name, department, salary FROM employees WHERE salary > 90000")

run_query("Task 1 - Sales or Finance",
          "SELECT * FROM employees WHERE department IN ('Sales','Finance')")


# Task 2

run_query("Task 2 - Top 5 Salaries",
          "SELECT * FROM employees ORDER BY salary DESC LIMIT 5")

run_query("Task 2 - Recent Engineering",
          "SELECT * FROM employees WHERE department = 'Engineering' ORDER BY hire_date DESC LIMIT 3")

run_query("Task 2 - Oldest (조건)",
          "SELECT * FROM employees WHERE salary > 70000 AND department <> 'HR' ORDER BY hire_date ASC LIMIT 1")


# Task 3

run_query("Task 3 - Monthly Salary",
          "SELECT name, salary, salary/12 FROM employees WHERE salary > 60000")

run_query("Task 3 - Finance Report",
          "SELECT name, salary, salary/12 FROM employees WHERE department = 'Finance' ORDER BY salary DESC")

# Close connection
conn.close()
```
